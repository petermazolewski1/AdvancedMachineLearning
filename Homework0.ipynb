```python
sf_rent = 1800
sf_utilities = 150
sf_food = 400
sf_transportation = 400
sf_entertainment = 400

sf_total_monthly = sf_rent + sf_utilities + sf_food + sf_transportation + sf_entertainment

sf_total_yearly = sf_total_monthly * 12

sd_rent = 1400
sd_utilities = 150
sd_food = 350
sd_transportation = 300
sd_entertainment = 350

sd_total_monthly = sd_rent + sd_utilities + sd_food + sd_transportation + sd_entertainment

sd_total_yearly = sd_total_monthly * 12

Seattle_rent = 1200
Seattle_utilities = 100
Seattle_food = 300
Seattle_transportation = 300
Seattle_entertainment = 300

Seattle_total_monthly = Seattle_rent + Seattle_utilities + Seattle_food + Seattle_transportation + Seattle_entertainment

Seattle_total_yearly = Seattle_total_monthly * 12
```

```python
import pandas as pd

df = pd.DataFrame(
    {
        "City": ["San Francisco", "San Diego", "Seattle"],
        "Rent": [sf_rent, sd_rent, Seattle_rent],
        "Utilities": [sf_utilities, sd_utilities, Seattle_utilities],
        "Food": [sf_food, sd_food, Seattle_food],
        "Transportation": [sf_transportation, sd_transportation, Seattle_transportation],
        "Entertainment": [sf_entertainment, sd_entertainment, Seattle_entertainment],
        "Total Monthly": [sf_total_monthly, sd_total_monthly, Seattle_total_monthly],
        "Total Yearly": [sf_total_yearly, sd_total_yearly, Seattle_total_yearly]
    }
)

minimum_annual_salary_sf = sf_total_yearly * 1.25
minimum_annual_salary_sd = sd_total_yearly * 1.25
minimum_annual_salary_seattle = Seattle_total_yearly * 1.25

print(f"Minimum annual salary for San Francisco: ${minimum_annual_salary_sf:,}")
print(f"Minimum annual salary for San Diego: ${minimum_annual_salary_sd:,}")
print(f"Minimum annual salary for Seattle: ${minimum_annual_salary_seattle:,}")


desired_annual_salary = 100000

```

```python
from plotnine import *
ggplot(df, aes(x="City", y="Total Yearly", fill = "City")) + geom_bar(stat="identity") + theme_minimal() + labs(title="Total Yearly Expenses by City", x="City", y="Total Yearly Expenses")



```
