## Modeling Approach

For this assignment, I explored two gradient boosting algorithms: **LightGBM** and **CatBoost**. I began by building baseline models for both approaches to establish a performance benchmark.

The baseline **LightGBM** model achieved a **balanced accuracy of 0.9648**, while the **CatBoost** model achieved a slightly lower **balanced accuracy of 0.9638**. LightGBM was selected as the primary model for further tuning.


To better understand how gradient boosting parameters impact model performance, I tested three different hyperparameter configurations for LightGBM:

```python
param_grid_1 = {
    "learning_rate": 0.05,
    "n_estimators": 500,
    "num_leaves": 31,
    "max_depth": -1,
    "min_child_samples": 20,
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    "random_state": 42
}

param_grid_2 = {
    "learning_rate": 0.03,
    "n_estimators": 300,
    "num_leaves": 63,
    "max_depth": -1,
    "min_child_samples": 20,
    "random_state": 42
}

param_grid_3 = {
    "learning_rate": 0.02,
    "n_estimators": 500,
    "num_leaves": 31,
    "max_depth": -1,
    "min_child_samples": 20,
    "random_state": 42
}
```


Param_grid_1 performed the best, achieving a balanced accuracy of 0.9648, which was the same as the baseline model.

Learning Rate:
The model with the highest learning rate (0.05) performed the best. As the learning rate decreased (0.03 → 0.02), performance slightly declined. This suggests that a moderately higher learning rate allowed the model to learn more effectively within the given number of estimators.
Number of Leaves (num_leaves):
Increasing the number of leaves from 31 to 63 (param_grid_2) did not improve performance. This indicates that increasing model complexity may have led to slight overfitting or diminishing returns.
Number of Estimators:
Param_grid_2 used fewer estimators (300), which may have limited its ability to fully learn patterns, especially with a lower learning rate. It is likely that increasing the number of estimators to 500 for this configuration could improve performance.
Subsampling:
Param_grid_1 included subsampling (subsample=0.8, colsample_bytree=0.8), which may have helped reduce overfitting and contributed to its stronger performance.
Max Depth:
The parameter max_depth = -1 allowed trees to grow without a strict depth limit, with complexity instead controlled by num_leaves.
