## 3. Model Approaches

https://github.com/petermazolewski1/AdvancedMachineLearning/blob/main/KagglePlaygroundS6E4/KaggleBaggingBoosting.ipynb

I used two modeling approaches to compare bagging vs. boosting:

Random Forest 
XGBoost

Both models were trained using the same encoded dataset and scored using balanced accuracy. I also performed light hyperparameter tuning using GridSearchCV with StratifiedKFold cross-validation.

Additionally, I used feature engineering including interaction terms and squared variables based on feature importance insights from earlier EDA.

Model Performance

Both models performed well overall and produced strong predictions. The confusion matrices showed that most predictions were correct, with relatively few misclassifications between irrigation levels.

Tuning results showed:

Random Forest improved slightly after tuning. It went from a 0.9532 to a 0.9589
XGBoost saw very minimal improvement after tuning. It went from a 0.961820 to a 0.961873

This means that XGBoost was already performing well with default parameters, while Random Forest benefited slightly from tuning.

### Best Score

My best model score was the xgboost with a balanced accuracy of 0.95926 which puts me in 1346th place.




## 4. Bagging vs Boosting

Boosting (XGBoost) consistently outdid bagging (Random Forest) in both cross-validation and leaderboard performance. XGBoost had slightly fewer classification errors in the confusion matrix and achieved higher balanced accuracy.

Random Forest performed well but did not do as good as XGBoost. Random Forest improved a little after tuning but still lagged behind XGBoost.
